```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import cv2
import os
import glob
import sys
sys.path.append('../')
from utils.Utility import ageEncoder
```

```python
train_df_path = '../../input/data/train/train.csv'
train_df = pd.read_csv(train_df_path)
train_df.sample(5)
```

```python
plt.figure(figsize=(15, 8))
sns.countplot(
    data=train_df,
    x='age',
    hue='gender'
)
plt.show()
```

```python
plt.figure(figsize=(15, 8))

comp_age_df = train_df.copy()
comp_age_df['age'] = comp_age_df['age'].apply(ageEncoder)

ax = sns.countplot(
    data=comp_age_df,
    x='age',
    hue='gender'
)

for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', xy=(p.get_x()+p.get_width()/2, p.get_height()+10), ha='center')
ax.set_xticklabels(['18~30', '30~60', '60~'])

plt.show()
```
